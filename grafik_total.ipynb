```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv("/content/statistik_mitigasi_kecamatan.csv")
```

```python
# Sort dari terbesar
df_sorted = df.sort_values("luas_mitigasi_ha", ascending=False)
```

```python
# Buat grafik
plt.figure(figsize=(12, 6))
plt.barh(df_sorted["nama_kec"], df_sorted["luas_mitigasi_ha"], color='forestgreen')
plt.xlabel("Luas Kehilangan (ha)")
plt.title("Total Kehilangan Tutupan Pohon per Kabupaten di Provinsi Jambi")
plt.gca().invert_yaxis()  # Kabupaten terbesar di atas
plt.tight_layout()
plt.savefig("/content/statistik_mitigasi_kecamatan.png")
plt.show()
```
